53. **Crie uma nova coluna para `df` chamada 'adjacent' que contenha o número de minas encontradas em quadrados adjacentes.**
```python
def count_adjacent_mines(row):
    adj_coords = [(row['x'] + dx, row['y'] + dy) for dx in [-1, 0, 1] for dy in [-1, 0, 1] if (dx, dy) != (0, 0)]
    return df.set_index(['x', 'y']).loc[adj_coords, 'mine'].sum()
df['adjacent'] = df.apply(count_adjacent_mines, axis=1)
print(df)
```

59. **Gere dados de ações aleatórios para um dia e agregue-os por hora.**
```python
import numpy as np
import pandas as pd
def float_to_time(x):
    return str(int(x)) + ":" + str(int(x%1 * 60)).zfill(2) + ":" + str(int(x*60 % 1 * 60)).zfill(2)
def day_stock_data():
    time = 9.5
    price = 100
    results = [(float_to_time(time), price)]
    while time < 16:
        elapsed = np.random.exponential(.001)
        time += elapsed
        if time > 16:
            break
        price_diff = np.random.uniform(.999, 1.001)
        price *= price_diff
        results.append((float_to_time(time), price))
    df = pd.DataFrame(results, columns = ['time','price'])
    df.time = pd.to_datetime(df.time)
    return df
df = day_stock_data()
df = df.set_index('time').resample('H').ohlc()
print(df)
```
Usamos `resample` com `ohlc` para agregar os dados por hora.

60. **Crie um gráfico de velas para os dados de ações.**
```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from mpl_finance import candlestick_ohlc
def plot_candlestick(agg):
    fig, ax = plt.subplots()
    df = agg.reset_index()
    df['time'] = df['time'].apply(mdates.date2num)
    candlestick_ohlc(ax, df.values, width=0.6, colorup='g', colordown='r')
    ax.xaxis_date()
    plt.title('OHLC of Stock Value During Trading Day')
    plt.xlabel('Time')
    plt.ylabel('Price')
    plt.show()
df = day_stock_data()
agg = df.set_index
('time').resample('H').ohlc()
plot_candlestick(agg)
```
Usamos `candlestick_ohlc` do `mpl_finance` para criar um gráfico de velas.
